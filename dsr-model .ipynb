```r
library(tidyverse)
library(Metrics)
```

```r
data <- read.csv("/kaggle/input/wine-quality-prediction/train.csv")
head(data)
dim(data)
```

```r
features <- c("fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", "total sulfur dioxide", "density", "pH", "sulphates", "alcohol")
target <- "quality"
```

```r
set.seed(9123)  # Để đảm bảo kết quả có thể tái lặp
train_indices <- sample(1:nrow(data), 0.95 * nrow(data))  # 70% dữ liệu cho tập huấn luyện
train_data <- data[train_indices, ]
test_data <- data[-train_indices, ]
```

```r
library(e1071)
# model <- lm(quality ~ ., data = train_data) linear regression
model <- svm(quality ~ ., data = train_data, kernel = "linear")

```

```r
summary(model)
```

```r
predictions <- predict(model, newdata = test_data)
```

```r
mae <- mean(abs(predictions - test_data$quality))  # Sai số tuyệt đối trung bình
rmse <- sqrt(mean((predictions - test_data$quality)^2))  # Sai số bình phương trung bình

print(paste("MAE:", mae))
print(paste("RMSE:", rmse))
```

```r
test_data1 <- read.csv("/kaggle/input/wine-quality-prediction/test.csv")
predictions <- predict(model, newdata = test_data1)
rounded_predictions <- round(predictions, digits = 0)
rounded_predictions <- pmin(pmax(round(predictions), 0), 10)
```

```r
result <- data.frame(id = test_data1$id, quality = predictions)
```
